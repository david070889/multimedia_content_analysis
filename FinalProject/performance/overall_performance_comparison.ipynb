```python
import matplotlib.pyplot as plt

# txt
def read_data(file_path):
    train_losses = []
    train_accuracies = []
    val_losses = []
    val_accuracies = []
    with open(file_path, 'r') as f:
        lines = f.readlines()
        i = 0
        while i < len(lines):
                if 0 < i < 26:
                    train_losses.append(float(lines[i].strip()))
                    i += 1

                elif 26 < i < 52:
                    train_accuracies.append(float(lines[i].strip()))
                    i += 1
                    
                elif 52 < i < 78:
                    val_losses.append(float(lines[i].strip()))
                    i += 1
                    
                elif 78 < i < 104:
                    val_accuracies.append(float(lines[i].strip()))
                    i += 1
                else:
                    i += 1
    return train_losses, train_accuracies, val_losses, val_accuracies

# 
train_losses_1, train_accuracies_1, val_losses_1, val_accuracies_1 = read_data('mobilenet.txt')
train_losses_2, train_accuracies_2, val_losses_2, val_accuracies_2 = read_data('shufflenetV2.txt')
train_losses_3, train_accuracies_3, val_losses_3, val_accuracies_3 = read_data('efficientnet_b0.txt')
train_losses_4, train_accuracies_4, val_losses_4, val_accuracies_4 = read_data('efficientformerV2_s0.txt')
train_losses_5, train_accuracies_5, val_losses_5, val_accuracies_5 = read_data('repvgg.txt')
train_losses_6, train_accuracies_6, val_losses_6, val_accuracies_6 = read_data('repvit.txt')


x = []
for i in range(25):
    x.append(i)

# 
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x, train_losses_1, label='mobilenetV3')
plt.plot(x, train_losses_2, label='shufflenetV2')
plt.plot(x, train_losses_3, label='efficientnet_b0')
plt.plot(x, train_losses_4, label='efficientformerV2_s0')
plt.plot(x, train_losses_5, label='repvgg')
plt.plot(x, train_losses_6, label='repvit')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train Loss')
plt.legend()

# 
plt.subplot(2, 2, 2)
plt.plot(x, val_losses_1, label='mobilenetV3')
plt.plot(x, val_losses_2, label='shufflenetV2')
plt.plot(x, val_losses_3, label='efficientnet_b0')
plt.plot(x, val_losses_4, label='efficientformerV2_s0')
plt.plot(x, val_losses_5, label='repvgg')
plt.plot(x, val_losses_6, label='repvit')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Val Loss')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, train_accuracies_1, label='mobilenetV3')
plt.plot(x, train_accuracies_2, label='shufflenetV2')
plt.plot(x, train_accuracies_3, label='efficientnet_b0')
plt.plot(x, train_accuracies_4, label='efficientformerV2_s0')
plt.plot(x, train_accuracies_5, label='repvgg')
plt.plot(x, train_accuracies_6, label='repvit')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train Acc')
plt.legend()

# 
plt.subplot(2, 2, 4)
plt.plot(x, val_accuracies_1, label='mobilenetV3')
plt.plot(x, val_accuracies_2, label='shufflenetV2')
plt.plot(x, val_accuracies_3, label='efficientnet_b0')
plt.plot(x, val_accuracies_4, label='efficientformerV2_s0')
plt.plot(x, val_accuracies_5, label='repvgg')
plt.plot(x, val_accuracies_6, label='repvit')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Val Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
