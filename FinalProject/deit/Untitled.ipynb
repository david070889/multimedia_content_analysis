```python
import timm
import torch
import torch.nn as nn

# 設定設備
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("using {} device.".format(device))

# 創建模型，預訓練且將分類層設置為2類
model = timm.create_model('deit_tiny_distilled_patch16_224', pretrained=False)
checkpoint = torch.load('deit-tiny-distilled_3rdparty_pt-4xb256_in1k_20211216-c429839a.pth', map_location=device)
model.head = nn.Linear(model.head.in_features, 2)
model.head_dist = nn.Linear(model.head_dist.in_features, 2)
# 刪除分類層權重
pre_dict = {k: v for k, v in checkpoint.items() if k in model.state_dict() and model.state_dict()[k].numel() == v.numel()}
missing_keys, unexpected_keys = model.load_state_dict(pre_dict, strict=False)

# 檢查模型架構
print(model)
```
